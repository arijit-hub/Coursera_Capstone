```python
#Importing the necessary libraries

import pandas as pd
import numpy as np
import geocoder
import folium
```

# Step 1: Importing Dataset

```python
# We are going to import the csv file created in the first part

df = pd.read_csv('Neighbourhood_data') # reading the csv file

df.head() #Displaying the first five rows
```

```python
geo_data = pd.read_csv('http://cocl.us/Geospatial_data') #The Geocoder didn't work properly hence collected the data from csv 
                                                         #file

geo_data.head() #Displaying the first five rows
```

```python
geo_data.shape #Checking the shape of the data
```

```python
df_toronto = pd.merge(df,geo_data,on="Postal Code") #merginf the two dataframes based on Postal Code

df_toronto.head(10) #Displaying the first 10 rows
```

```python
df_toronto.shape #Checking the shape to be certain
```

# Step 3: Data  Visualization

```python
toronto_lat = 43.6532 #Defining latitude of Toronto
toronto_long = -79.3832 #Defining longitude of Toronto
```

```python
toronto_map = folium.Map(location = [toronto_lat,toronto_long],zoom_start = 10) #Plotting the map of Toronto


#Plotting neighbourhoods

for lat,long,borough,neighbourhood in zip(df_toronto['Latitude'],df_toronto['Longitude'],df_toronto['Borough'],df_toronto['Neighbourhood']):
    
    label = '{}, {}'.format(neighbourhood, borough) #Labelling the neighbourhood
    label = folium.Popup(label, parse_html=True)
    
    #Marking the circle
    folium.CircleMarker(location=[lat,long], 
                        radius=5,
                        popup=label,
                        color='#5F34A3',
                        fill=True,
                        fill_color='#00A9A1',
                        fill_opacity=0.7,
                        parse_html=False).add_to(toronto_map)

toronto_map
```
